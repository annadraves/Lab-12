```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X = ['In Air','Room Temperature Water','Ice Water']
dom = [-48.9,-20.0,-26.2]
```

```python
X_axis = np.arange(len(X))
  
plt.bar(X_axis, dom, label = 'Condition of Apnea')
  
plt.xticks(X_axis, X)
plt.xlabel("Conditions")
plt.ylabel("Percent Change from Resting HR")
plt.title("Conditions of Apnea")
plt.legend()
plt.show()
```
